```python

#file= Zx_5.ipynb
#Author= Abdulla Al-Jabri
#Sana'a- Yemen
#Jabri62018@gmail.com
#May 8, 2026 AD

import numpy as np

# 1. The 5 numbers - no physics here
zeros = np.array([14.134725, 21.022040, 25.010858, 30.424876, 32.935062])

# 2. Your Z(x) function - the only formula you need
def Zx(x, xp=32.935062):  # xp = 5th zero
    return x**5 * np.log(x) * np.sin(2*np.pi/x) * np.exp(-x/xp)

# 3. Play: plug the first 4 zeros into Z(x)
Z_values = [Zx(t) for t in zeros[:4]]

# 4. What if? Combine them to get physics
Z = np.mean(Z_values)           # Geometric time
C = np.std(Z_values)            # Curvature
A = np.max(Z_values) - Z - C    # Structural acceleration

Z_t = Z + C + A
print(f"Z_t = Z + C + A = {Z_t:.6f}")
print(f"Expected: 1.000000")

# 5. Now derive stuff:
H0_predicted = 67.4 * Z_t       # km/s/Mpc
w_predicted = -1.0 - A/Z_t      # dark energy equation of state

print(f"H0 = {H0_predicted:.1f} km/s/Mpc  ¡ú Planck: 67.4")
print(f"w  = {w_predicted:.3f}         ¡ú DESI 2024: -1.03")
```
